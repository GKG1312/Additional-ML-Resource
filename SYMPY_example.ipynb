Importing all the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import sympy as smp 
from sympy import *
from sympy import Eq, Symbol, solve
from scipy import stats
import math as mt
```

writing expression an plotting curves

```python
x=Symbol('x')
y=Symbol('y')

y=1/(2*smp.sqrt(2*np.pi)) * smp.exp(-((x-0)**2)/(2*2**2))
print(y)
dy=y.diff(x)
print(dy)
s=solve(dy)
print(s)
ddy=dy.diff(x)
print(ddy)
z=ddy.subs(x,s[0])
print(z)
if z<0:
  print('\n Maxima found at : ',s,' and the maximum value is ',y.subs(x,s[0]))
print('\n Saddle point = ',s)
smp.plotting.plot(y)
smp.plotting.plot(dy)
smp.plotting.plot(ddy) 
```
